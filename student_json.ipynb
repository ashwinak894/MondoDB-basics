```python
#1. Find the student name who scored maximum scores in all (exam, quiz and homework)?
```

```python
x = col.aggregate([{'$unwind':'$scores'},{'$group':{'_id':'$_id','name':{'$first':'$name'},'max_sco':{'$sum':'$scores.score'}}}])
max = 0
for i in x:
  if i['max_sco']>max:
    max = i['max_sco']
    name = i['name']

print(name,max)
```

```python
#2. Find students who scored below average in the exam and pass mark is 40%?
```

```python
y = col.aggregate([{'$unwind':'$scores'},{'$group':{'_id':'$_id','name':{'$first':'$name'},'max_sco':{'$sum':'$scores.score'}}},{'$addFields':{'prcnt':{'$divide':['$max_sco',3]}}}])
for i in y:
  if i['prcnt']>=40 and i['prcnt']<=55:
    print(i)
```

```python
# 3.Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.
```

```python
z = col.aggregate([{'$unwind':'$scores'},
                           {'$group':{"_id":"$_id","name":{"$first":"$name"},"max_sco":{'$sum':"$scores.score"}}},
                           {'$addFields':{"prcnt":{'$divide':["$max_sco",3]}}},
                            {'$addFields':{"result":{"$cond":{'if' :{"$gte":["$prcnt",40]},'then':"Pass",'else':"Fail"}}}}])

for i in z:
  print(i['name'],'\n',i['result'])
```
